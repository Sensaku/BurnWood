# Quels sont les échantillons de la base de données SACHA ?

```sparql
select ?sample_name ?label_taxon where {
    ?sample skos:prefLabel ?sample_name;
          pl_tax:identifiedTaxon ?label.
    
    ?colletion skos:member ?sample;
             skos:prefLabel "SACHA".
    
    ?taxon txrfp:hasReferenceName ?label.
    ?label skos:prefLabel ?label_taxon.
}
```

## Quels sont les individus de la base de données SACHA se trouvant en "Northern Provinces" ?

```sparql
select ?label_taxon ?sample_name ?loc_id ?loc_name where {
  ?sample skos:prefLabel ?sample_name;
	pl_tax:identifiedTaxon ?label.
  ?colletion skos:member ?sample;
             skos:prefLabel "SACHA".
   
  ?taxon txrfp:hasReferenceName ?label;
         wcvp:location ?location2.

  ?label skos:prefLabel ?label_taxon.

  ?location2 a wcvp:Area;
               rdfs:label "TVL";
            skos:prefLabel ?loc_name.
}
ORDER BY ?sample_name
```

# Quels sont les individus de bois de la base de données SACHA qui ne sont pas en "Northen Provinces" ?

```sparql
select ?label_taxon ?sample_name where {
?sample skos:prefLabel ?sample_name;
        pl_tax:identifiedTaxon ?label.
    
?colletion skos:member ?sample;
           skos:prefLabel "SACHA".

?taxon txrfp:hasReferenceName ?label.
?label skos:prefLabel ?label_taxon.
    
MINUS {
    
?taxon txrfp:hasReferenceName ?label;
       wcvp:location ?location2.

?label skos:prefLabel ?label_taxon.

?location2 a wcvp:Area;
           rdfs:label "TVL";
           skos:prefLabel ?loc_name.
     
 }
  
}
ORDER BY ?sample_name
```
