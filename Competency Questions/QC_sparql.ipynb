# Quels sont les échantillons de la base de données SACHA ?

```sparql
select ?sample_name ?label_taxon where {
    ?sample skos:prefLabel ?sample_name;
          pl_tax:identifiedTaxon ?label.
    
    ?colletion skos:member ?sample;
             skos:prefLabel "SACHA".
    
    ?taxon txrfp:hasReferenceName ?label.
    ?label skos:prefLabel ?label_taxon.
}
```

## Quels sont les individus de la base de données SACHA se trouvant en "Northern Provinces" ?

```sparql
select distinct ?label_taxon ?sample_name ?loc_id ?loc_name where {
  ?sample skos:prefLabel ?sample_name;
	pl_tax:identifiedTaxon ?label.
  ?colletion skos:member ?sample;
             skos:prefLabel "SACHA".
   
  ?taxon txrfp:hasReferenceName ?label;
         wcvp:location ?location2.

  ?label skos:prefLabel ?label_taxon.

  ?location2 a wcvp:Area;
               rdfs:label "TVL";
            skos:prefLabel ?loc_name.
}
ORDER BY ?sample_name
```

# Quels sont les individus de bois de la base de données SACHA qui ne sont pas en "Northen Provinces" ?

```sparql
select distinct ?label_taxon ?sample_name where {
?sample skos:prefLabel ?sample_name;
        pl_tax:identifiedTaxon ?label.
    
?colletion skos:member ?sample;
           skos:prefLabel "SACHA".

?taxon txrfp:hasReferenceName ?label.
?label skos:prefLabel ?label_taxon.
    
MINUS {
    
?taxon txrfp:hasReferenceName ?label;
       wcvp:location ?location2.

?label skos:prefLabel ?label_taxon.

?location2 a wcvp:Area;
           rdfs:label "TVL";
           skos:prefLabel ?loc_name.
     
 }
  
}
ORDER BY ?sample_name
```

# Wood identification with SPARQL (like Inside Wood mismatch feature research)

```sparql
prefix rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
prefix rdfs: <http://www.w3.org/2000/01/rdf-schema#>
prefix owl: <http://www.w3.org/2002/07/owl#>
prefix iawa:<https://www.iawa-website.org/core#>
prefix skos: <http://www.w3.org/2004/02/skos/core#>
prefix pl_tax: <http://www.plant-taxonomy-kg.com/core#>
prefix sosa: <http://www.w3.org/ns/sosa/>
prefix wcvp: <http://www.wcvp-core.com/class#>
prefix txrfp: <http://taxref.mnhn.fr/lod/property/>
# Add filter for max mismatch value (InsideWood at most 5 mismatche)
select ?sample ?family_name ?label ?mismatch ?origin where{
    ?classes txrfp:hasReferenceName ?taxon;
             rdfs:subClassOf
             rdfs:subClassOf/rdfs:subClassOf ?family.
    ?family rdfs:label ?family_name.
    # Find closest similar sample base on mismatch
    { 
        select ?sample ?taxon ?label ((?origin - COUNT(?feature)) as ?mismatch) ?origin where {
            #Found Sample feature
        	{
                SELECT ?feature WHERE {
        		<https://www.wood-sample-observation.com/sample/IW-1005> pl_tax:hasObservation [sosa:hasResult [rdf:value ?feature]].
                }
            }
            # Count Sample feature
        	{
                SELECT (COUNT(?feature) as ?origin) WHERE {
        		<https://www.wood-sample-observation.com/sample/IW-1005> pl_tax:hasObservation [sosa:hasResult [rdf:value ?feature]].
                }
            }
            ?sample pl_tax:hasObservation [ sosa:hasResult [ rdf:value ?feature]];
                pl_tax:identifiedTaxon ?taxon.
            ?taxon skos:prefLabel ?label.
            FILTER (?sample != <https://www.wood-sample-observation.com/sample/IW-1005>)
        }
        GROUP BY ?sample ?origin ?label ?taxon
        ORDER BY ?mismatch 
    }
FILTER (?mismatch <=5)
}
ORDER BY ?mismatch
```
